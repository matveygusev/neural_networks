После первичного анализа датасета можно сделать вывод, что задача заключается в анализе данных о клиентах. Целевой переменной в таком датасете является **"Response"**

## 3. Распределение целевой переменной

```python
plt.figure(figsize=(8, 6))
df['Response'].hist(bins=2, edgecolor='k')
plt.title('Распределение целевой переменной (Response)')
plt.xlabel('Response')
plt.ylabel('Количество')
plt.xticks([0, 1])
plt.grid(False)
plt.show()

unique_responses = df['Response'].nunique()
print(f"Количество уникальных значений целевой переменной: {unique_responses}")
```

## 6. Количество пропусков

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
mean = df['Income'].mean()
df['Income'].fillna(mean, inplace=True)
print('Кол-в пропусков после замены')
print(df.isnull().sum())
```

## 7. Гистограммы для 5 признаков

```python
plt.figure(figsize=(12, 8))
df.boxplot(column='Income', by='Education', grid=False)
plt.title('Boxplot of Income by Education')
plt.suptitle('')
plt.xlabel('Education')
plt.ylabel('Income')
plt.xticks(rotation=45)
plt.show()

features = ['Year_Birth', 'MntWines', 'NumWebPurchases', 'NumStorePurchases']
df[features].hist(bins=30, figsize=(15, 10), edgecolor='k')
plt.show()
```

## 8. Зависимости признаков от целевой перменной

```python
features_to_plot = ['Income', 'Year_Birth', 'MntWines', 'NumWebPurchases', 'NumStorePurchases']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_to_plot, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='Response', y=feature, data=df)
    plt.title(f'Boxplot of {feature} by Response')
plt.tight_layout()
plt.show()
```

**Вывод**: есть как полезные переменные вроде "Income" и "MntWines", так и бесполезные такие как "Year_birth" или "NumWebPurchase"
